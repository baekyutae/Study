```python
import threading
import time

# ì‹ í˜¸ë“± ìƒì„± (ì´ˆê¸° ìƒíƒœ: Red Light)
db_ready_event = threading.Event()

def connect_database():
    print(" [DB ìŠ¤ë ˆë“œ] ë°ì´í„°ë² ì´ìŠ¤ ì—°ê²° ì‹œë„ ì¤‘...")
    time.sleep(4) # ì—°ê²° ê³¼ì • ì‹œë®¬ë ˆì´ì…˜
    print(" [DB ìŠ¤ë ˆë“œ] ì—°ê²° ì„±ê³µ! ì»¤ë„¥ì…˜ í’€ ìƒì„± ì™„ë£Œ.")
    
    # [ê¹ƒë°œ ë“¤ê¸°] "ì¤€ë¹„ ëì–´!"ë¼ê³  ì‹ í˜¸ë¥¼ ë³´ëƒ„
    db_ready_event.set() 

def start_api_server():
    print(" [ì„œë²„ ìŠ¤ë ˆë“œ] ì„œë²„ êµ¬ë™ ì¤€ë¹„...")
    
    # [ëŒ€ê¸°] DBê°€ ì¤€ë¹„ë  ë•Œê¹Œì§€ ì—¬ê¸°ì„œ ë©ˆì¶¤
    # (joinì€ 'ì£½ì„ ë•Œ'ê¹Œì§€ ê¸°ë‹¤ë¦¬ì§€ë§Œ, ì´ê±´ 'ê¹ƒë°œ ë“¤ ë•Œ'ê¹Œì§€ë§Œ ê¸°ë‹¤ë¦¼)
    print(" [ì„œë²„ ìŠ¤ë ˆë“œ] DB ì—°ê²° ëŒ€ê¸° ì¤‘... (Blocking)")
    db_ready_event.wait()
    
    print("\nâœ… [ì„œë²„ ìŠ¤ë ˆë“œ] ì‹ í˜¸ ìˆ˜ì‹ ! 8080 í¬íŠ¸ë¡œ ì„œë²„ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤.")
    print("    í´ë¼ì´ì–¸íŠ¸ ìš”ì²­ ìˆ˜ë½ ê°€ëŠ¥ ìƒíƒœ")

# --- ì‹¤í–‰ ---
t1 = threading.Thread(target=connect_database)
t2 = threading.Thread(target=start_api_server)

t1.start()
t2.start()

# ì˜ˆì‹œë¼ì„œ ë©”ì¸ ìŠ¤ë ˆë“œ ëŒ€ê¸°ìš© join ì¶”ê°€ (ì‹¤ì œ ì„œë²„ëŠ” ê³„ì† ì¼œì ¸ ìˆìŒ)
t1.join()
t2.join()
```
