m은 항상 n 이하

key는 이동 가능

90도로 3번 => 3가지 경우의 수
상하좌우로 각각 2번씩 => 16개의 경우의 수(key와 lock의 접점이 존재하는 한해에서)

겹치는 영역에서 각각 lock은 0이고 key는 1이여야 함
