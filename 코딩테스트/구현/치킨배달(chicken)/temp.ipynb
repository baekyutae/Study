집에서 각 치킨집의 거리를 구하고 가장 작은 값이 그집의 치킨거리

도시의 치킨 거리의 최솟값

n개의 치킨집이 존재하고 그중에 최대 m개만 남김

일단 단 한번도 특정 집의 치킨거리가 되지 못한 치킨집을 폐점

0~m개 까지 페업 하는 경우의수


집을 고르고 가장 가까운 치킨집의 치킨거리를 구함
m개 제한을 지키는 선에서 도시의 치킨거리의 최솟값을 구함

```python
info = []
def chicken(N,M, info):
    집들을 순회 하며 선택
    선택된 집들에서 가장 가까운 치킨집을 찾음=> 치킨거리
    치킨집들의 딕셔너리를 만들어 치킨거리로 선정된 횟수를기록
    m을 기준삼아 치킨거리로 선정된 횟수가 적은 순으로 폐점
```

```python
존재하는 치킨집 중에서 m개를 고름

조합을 가지고 각 집들의 치킨 거리를 구하고 그합 즉 도시의 치킨 거리를 구함

가장 도시의 치킨 거리 값이 작은 경우를 찾음
```

```python
# 시간복잡도 : O( (C choose m) · H · m )

from itertools import combinations

n, m = map(int, input().split())
chicken, house = [], []
data = []
for r in range(n):
    for c in range(n):
        if data[c] == 1:
            house.append((r, c)) # 일반 집
        elif data[c] == 2:
            chicken.append((r, c)) # 치킨집

# 모든 치킨 집 중에서 m개의 치킨 집을 뽑는 조합 계산
candidates = list(combinations(chicken, m))

# 치킨 거리의 합을 계산하는 함수
def get_sum(candidate):
    result = 0
    # 모든 집에 대하여
    for hx, hy in house:
        # 가장 가까운 치킨 집을 찾기
        temp = 1e9
        for cx, cy in candidate:
            temp = min(temp, abs(hx - cx) + abs(hy - cy))
        # 가장 가까운 치킨 집까지의 거리를 더하기
        result += temp
    # 치킨 거리의 합 반환
    return result

# 치킨 거리의 합의 최소를 찾아 출력
result = 1e9
for candidate in candidates:
    result = min(result, get_sum(candidate))

print(result)
```
