N * N 공간

LRUD 로 이동

공간을 벗어나는 움직임은 무시

최종 도착 위치의 좌표

시작은항상 1,1
(X,Y)
L : (X-1,Y)
R : (X+1,Y)
U : (X,Y+1)
D : (X,Y-1)



```python
# 내 풀이


n = int(input())
move = list(input().split())

def location_detector(n):
    point = [1,1]
    for i in move:
        if i == "L":
            point[1]-=1
            if point[1] < 1:
                point[1] +=1
        elif i == "R":
            point[1]+=1
            if point[1] > n:
                point[1] -=1
        elif i == "U":
            point[0]-=1
            if point[0] < 1:
                point[0] +=1

        else: #i == "D"
            point[0]+=1
            if point[0] > n:
                point[0] -=1

    return point

location_detector(n)



            
```

```python
# CASE 1 : U방향으로 공간 밖으로 벗어나는 경우

5, R R R U D D, [3,4]

# CASE 2 : R, D, L 방향으로 한번씩 벗어나는 경우

5, R R R R R D D D D D L L L L L , [5,1]

# CASE 3 : 최소단위(한번)로 이동

5, R , [1,2]
```
