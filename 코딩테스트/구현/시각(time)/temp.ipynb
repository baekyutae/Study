정수 n

00시00분00부터 n시 59분 59초 => 0부터 n5959까지

의 모든 시각중에서 3이 하나라도 포함

일단 n시 59분 59초 까지 증가하며 3이 들어 있는지 여부 확인?...

일단 시계를 만들어야함

하루 86400초

```python
h = int(input())

def number_check(h):
    count = 0
    for i in range(0,h+1):
        for m in range(60):
            for s in range(60):
                if '3' in str(i)+str(m)+str(s):
                    count += 1
    return count

number_check(h)
```
