정수 n 이 주어짐 자릿수는 항상 짝수 

10이상 천만 미만

입력된 정수를 해체 해서 반반으로 나눠 각각의 자릿수 합을 구해야함

출력

스킬 사용 가능하면 lucky
안되면 ready

number를 문자 처럼 취급해 분리가 가능하게 만든 다음 배열에 담음

배열의 길이를 제고 절반값을 구해 인덱싱


두개의 배열로 분리 각각의 배열의 원소들을 int로 다시 변환하고 sum을 구해 비교

if문으로 결과에 따른 스킬 사용 여부 응답

```python
number = int(input())

def skill_check(number):
    point = []
    for i in str(number):
        point.append(i)

    mid = len(point)//2

    left_sum = sum(int(i) for i in point[:mid])
    right_sum = sum(int(i) for i in point[mid:])

    if left_sum == right_sum:
        print("lucky")
    else:
        print("ready")

skill_check(number)

# 시간복잡도는 O(n) 
#  자릿수의 길이만큼 for문 돌아감 => n
# 반으로 나눠서 sum  = > n/2 + n/2 == n
```
