출력: 압축했을 때 가장 짧은 것의 길이

```python
s = "ababcdcdefg"

temp = []
result = ""
zip_size = 2

while zip_size < len(s):
    for i in range(len(s)):
        temp.append(s[i])
        if len(temp) == zip_size*2 and temp[0:zip_size] == temp[zip_size:]:
            result += str(zip_size) + "".join(temp[:zip_size])
            temp = []
        else:
            result += s[i]
    print(result)
    result = ""
    zip_size += 1

# print(result)
```
