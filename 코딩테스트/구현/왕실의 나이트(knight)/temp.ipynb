8*8 좌표 평면

행은 1~8
열은 a~h

좌표를 나타내는 두문자로 구성된 문자열 입력 : a1(열행)

좌표에서 나이트가 이동할수 있는 경우의 수

1. 수평 두칸 수직 한칸 : 4가지
2. 수직 두칸 수평 한칸 : 4가지

이동 가능한 경우의 수 배열을 만들어 순회하면서 실제로 이동 가능한지 체크
또는 딕셔너리로 만들어 해당하는 value만큼 이동 시도

이동 안되는 조건

1. 수직 이동시 <1 or >8
2. 수평 이동시 a에서 좌측으로 이동 or h에서 우측으로 이동 


a~h를 숫자로 치환해서 

```python
data= input()

x,y = data[0], data[1]

change = {"a":1, "b":2, "c":3, "d":4, "e":5, "f":6, "g":7, "h":8}

x,y = change[x], int(y)


def knight_move(x,y):
    count = 0
    direction = {
    "LeftUP" : (-2,-1),
    "LeftDown": (-2,1),
    "RightUP" : (2,-1),
    "RightDown": (2,1),
    "UpLeft" : (-1,-2),
    "UpRright" : (1,-2),
    "DownLeft" : (-1,2),
    "DownRight" : (1,2)
    }
    for dir in direction:
        move_x, move_y = direction[dir]
        x, y = x + move_x, y + move_y
        #print(x,y)
        if x >0 and x < 9 and y > 0 and y < 9:
            #print(f"{dir} 방향으로 이동 가능")
            count += 1
        x,y = x - move_x, y - move_y # 처음 입력 자리로 복귀

    return count

knight_move(x,y)



       
```

```python
def solution():
    input_data = input() # 예: "a1"
    
    # 1. 문자를 숫자로 변환 (ord 사용)
    # row: 숫자 부분 (1~8)
    row = int(input_data[1])
    # col: 알파벳 부분 (a~h) -> (1~8)로 변환
    col = int(ord(input_data[0])) - int(ord('a')) + 1

    # 2. 8가지 방향 정의 (리스트가 딕셔너리보다 순회가 간편함)
    steps = [
        (-2, -1), (-2, 1), (2, -1), (2, 1), 
        (1, -2), (1, 2), (-1, -2), (-1, 2)
    ]

    count = 0

    for step in steps:
        # 3. 임시 변수(next)로 이동할 위치 미리 계산
        next_row = row + step[0]
        next_col = col + step[1]

        # 4. 경계값 체크 (1~8 사이인지)
        if 1 <= next_row <= 8 and 1 <= next_col <= 8:
            count += 1
            
    print(count)

# 실행
solution()
```
