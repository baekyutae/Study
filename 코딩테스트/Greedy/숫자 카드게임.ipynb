생각

가장 높은 숫자가 쓰인 카드를 뽑는게 최종 목적

카드는 2차원 배열로 배치

행을 고르고 해당 행에서 열을 골라 카드를 선택

단

행 안에서 숫자가 가장 작은 카드를 뽑아야함

즉 행의 최소값들을 비교해 그중 최대값을 뽑아야함

<입력>

첫째줄 : 행의 갯수 N 열의 갯수 M
둘째줄부터 N개줄 : 각 행별 카드에 적힌 숫자 

행은 1이상 열은 100이하

<출력>
제일 큰수

----

리스트 형태로 카드 숫자를 받아 인덱스 슬라이싱 하여 행을 구분 하려함

a = [3 1 2 4 1 4 2 2 ]
a[:N]으로 끊으면 하나의 행이 됨
start = 0
end = N
temp = []
final=[]

while len(final) < len(a) // M
    for i in a[start:end]:
        temp.append(i)

    start += M
    end += M
    final.append(min(temp))
    temp = [] # 초기화

이러면 마지막 행의 카드 개수가 M개가 아니면 입력 배열에서 문제가 생길거 같음 : end에서 out of range


아니면 입력할때 [[3,1,2], [4,1,4],[2,2]] 이렇게 받으면?

그러면 순회 하면서 각 내부 리스트에 대해 min() 해서 따로 만든 리스트에 append?

```python


N, M = map(int, input().split())
array_list = []


for i in range(0, N):
    array_list.append(list(map(int, input().split())))

def card_game(N,M,array_list):
    compare_list = []
    for temp in array_list:
        compare_list.append(min(temp))
    
    print(max(compare_list))

card_game(N,M,array_list)

# 입력 읽기: O(N*M)
# 각 행의 min: O(N*M)
# 전체는 여전히 O(N*M) (상수 배는 무시)
```
