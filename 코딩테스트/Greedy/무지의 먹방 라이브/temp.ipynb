회전판 

1초에 하나씩 음식이 돌아가며 바뀜

food_times 는 음식별 다먹기 위해 걸리는 시간 배열

k초에서 장애 발생 할때까지 회전판 돌아가며 음식 먹어야함

for문은 일방향이니 처음 으로 다시 돌아오기 힘듬

```python
# 1초씩 체크 하는 방식

# sum(food_times) <= k면 -1 처리해야 함
# 문제 조건상 k초 전에 음식이 다 떨어지면 -1인데, 그 케이스 처리가 없다.
    
time = 0
k=5
food_times = [3,1,2]
while time < k:
    for i in range(0,len(food_times)):
        print("for문 시작 현재 시각",time)
        print("현재 index", i)
        if food_times[i] == 0:
            if k == time:
                print("장애 발생1")
                if food_times[-1] == food_times[i]: 
                    print(1)
                else:
                    print(i+2) # => i+2를 해서 다먹은 음식이 선택되면 문제 발생
                break
            else:
                time+=1 # => 다먹은 음식은 시간 소요가 안됨 수정 필요
                print("한 음식 다먹었고 아직 장애 발생전")
            
        else:
            if k == time:
                print("장애 발생2")
                if food_times[-1] == food_times[i]: # 처음 음식 부터 재개 할 경우
                    print(1) 
                else:
                    print(i+2) # index 0,1,2니깐 +1 그리고 먹은 접시의 다음 접시니깐 +1
                
                break
            else:
                food_times[i] -= 1
                print("1초 먹음",food_times)
                time+=1


    


```
