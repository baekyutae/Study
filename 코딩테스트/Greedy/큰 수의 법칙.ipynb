```python
# 수정 2

N, M, K = map(int, input().split())
num_list = list(map(int, input().split()))

def big_Num_rule(N, M, K):
    sum_list = []
    
    num_list.sort(reverse=True)
    # print(num_list)
    if N == 1:
        return print(num_list[0] * M)
    
    if num_list[0] == num_list[1]:
        return print(num_list[0] * M)
    else:
        while len(sum_list) < M:
            for _ in range(0,K):
                if len(sum_list) < M:
                    sum_list.append(num_list[0])
            if len(sum_list) < M:
                sum_list.append(num_list[1])

        return print(sum(sum_list))
            


big_Num_rule(
    N,M,K
)

# 정렬: O(N log N)
# 합 계산: 최악 M번 append → O(M)
# 총합: O(N log N + M), 공간도 O(M) (sum_list 때문에)
```

```python
# 시간복잡도 O(N log N)인 방식

N, M, K = map(int, input().split())
num_list = list(map(int, input().split()))

def big_Num_rule(N, M, K):
    num_list.sort(reverse=True)
    if N == 1:
        return print(num_list[0] * M)

    a = num_list[0]
    b = num_list[1]

    if a == b:
        return print(a * M)
    # block : 제일 큰수 k개 + 두번째로 큰수 1개가 한묶음 
    block = K + 1
    full_blocks = M // block # 배열에서 묶음이 몇개 나오는지
    remainder = M % block # 묶음 제외한 자투리 개수

    count_a = full_blocks * K + remainder # 제일 큰수의 갯수
    count_b = full_blocks # 두번째로 큰수의 갯수: 묶음에 두번째로 큰수는 하나씩 들어가니 들어갈 수 있는 묶음 개수랑 동일 

    return print(count_a * a + count_b * b)

big_Num_rule(N, M, K)
```
