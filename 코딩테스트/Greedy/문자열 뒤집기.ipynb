일단 확실치는 않지만 1과 0의 갯수가 다를 때 더적은 걸 바꾸는게 행동 횟수가 적을 것 같음 

연속된 숫자만 바꿀 수 있고 사이에 다른수가 끼면 다음 행동에서 바꿀수 있게 구현해야함

```python
S = input()

zero_count = 0
one_count = 0
for i in S:
    i = int(i)
    if i == 1:
        one_count += 1
    else:
        zero_count += 1

if one_count > zero_count:
    for i in S:
        if i == 0:
            i = 1
            
```

아이디어가 잘못됨 : 1과 0 개별의 갯수 비교가 아닌

0으로 이루어진 연속 구간 개수와 1로 이루어진 연속 구간 개수를 세고, 그 중 더 작은 값이 최소 뒤집기 횟수.

1001001011 : 1 5개 0 5개

도 그룹으로 보면 1: 4그룹 0: 3그룹 즉 0을 1로 바꾸는게 최소 행동 


```python
# 시간복잡도 O(N)

S = input()

zero_group = 0
one_group = 0

# 첫 문자로 첫 그룹 시작
if S[0] == '0':
    zero_group += 1
else:
    one_group += 1

# 이전 문자와 달라질 때마다 새 그룹 시작
for i in range(1, len(S)):
    if S[i] != S[i-1]:
        if S[i] == '0':
            zero_group += 1
        else:
            one_group += 1

print(min(zero_group, one_group))
```

```python
# 다른 풀이
s=input()
cnt=0
for i in range(len(s)-1):
    if s[i]!=s[i+1]:
        cnt+=1 # 0과 1 사이 변경 횟수 +  1 이 총 그룹 갯수
    
print((cnt+1)//2)
```
