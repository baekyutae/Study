0부터 9로만  이루어진 문자열 s

사이에 * 또는 + 연산자를 넣어 만들 수 있는 가장 큰수를 구해라

연산은 항상 왼쪽부터 이루어짐

또한 가장큰 수는 항상 20억 이하의 정수가 되도록 입력이 주어짐
----

일단 문자열이니 각 문자를 분리하고 숫자로 바꿔 리스트에 저장 하여 계산이 가능하게 만듬

0이 오는 경우를 제외하곤 모두 곱을 하는게 가장 큰수가 나옴

result는 0

0 2 9 8 4 

0을 곱해봄 =>0 

더하기 0을 함

result 는 0

2를 곱해봄 => 0

더하기 2를 함

result 는 2

9를 곱해봄 => 19

result 18
---

2 9 0 8 4 

result 는 0

2를 곱해봄 => 0

더하기 2를함 

result 는 2

9를 곱해봄 =>18

result 는 18

0을 곱해봄 => 0

더하기 0을 함

result는 18



```python
n_string = input()
cal_list = []

for s in n_string:
    cal_list.append(int(s))

print(cal_list)

result = 0

for i in cal_list:
    if result * i == 0:
        result += i
    else:
        result *= i

print(result)
# 0 2 9 8 4
```

문제가 있음: 숫자가 0인 경우만 예외를 처리 했고 1인 경우는 곱해버려서 오히려 손해임 

그리고 result가 1인경우 그다음으로 오는 숫자의 곱 보다 합이 이득임 : 1+5 > 1*5

```python
# 시간복잡도 O(n) => cal_list 길이 N

n_string = input()
cal_list = []

for s in n_string:
    cal_list.append(int(s))

print(cal_list)

result = 0

for i in cal_list:
    if i <= 1 or result <= 1:
        result += i

    else:
        result *= i

print(result)
```
