1. n에서 1을 빼기

2. n을 k로 나누기 (k로 나누어 떨어질 때만 가능)

두가지 선택지를 반복해 1이 될떄 까지 진행

1번 혹은 2번 과정을 수행해야 하는 최소 횟수를 구해야함

입력시 n은 항상 k보다 크거나 같음 n >= k

----

일단 1로 빠르게 접근 하려면 1을 빼는 횟수보다 나누기를 더 많이 해야할 것 같음

n이 k로 나누어 나머지가 없는지 확인

나누기

안되면 1빼고 다시 K로 나눠지는지 확인

while n > 1 으로 반복

나누기와 1 뺴고 나누기에 대해 count += 1해주기

```python
N, K = map(int, input().split())

def divide_to_one(N,K):
    count = 0
    while N > 1:
        if N % K == 0:
            N = N//K
            count += 1
        else:
            N = N - 1
            count += 1

    print(count)

divide_to_one(N,K)

# 시간복잡도 O(N)

# N이 K로 나눠지지 않을 때는 매번 N = N - 1만 실행.
# 최악의 경우 거의 매번 1씩 줄어들어서 N이 1 될 때까지 약 N-1번 반복함.
# 따라서 반복 횟수는 N에 비례해서 O(N)이 됨.
```

```python
# 다른 풀이
# 시간 복잡도 O(logN) (밑이 K)

n, k = map(int,input().split())
 
count = 0
while True:
  target = (n//k)*k # n이 k로 나누어 떨어질 수 있는 target 찾기
  count += (n - target) # n에서 target 까지 되도록 -1 해야하는 횟수 더함
  n = target 
 
  if n < k: # n이 더이상 k로 나눠질 수 없다면 빠져나감
    break
    
  count += 1 # n을 k로 나누기
  n //= k
  
count += (n-1) # 남은 수에서 1이 될때까지 빼는 횟수 더함
print(count)

# 기존의 내 풀이는 매번 1을 빼면서 k의 배수인지 확인을 해야 했으나 이방식으로 하면 그 과정을 생략할수 있다.
# (n//k)*k 으로 나누어 떨어지는 수를 찾은 다음 n - target 하여   몇번 1을 빼줘야 하는지 한번에 구함
```

```python
# 동일한 로직인데 divmod 사용

n, k = map(int, input().split())

count = 0
while n >= k:
    q, r = divmod(n, k)   # n을 k로 나눈 몫, 나머지를 각각 q, r에 할당
    count += r            # r번 만큼 -1 한 것과 동일
    n = q                 # 나눠버림 (실제로는 n -= r 후 n//=k)
    count += 1            # 나누기 1회

count += (n - 1)          # 남은 건 1까지 -1
print(count)
```
