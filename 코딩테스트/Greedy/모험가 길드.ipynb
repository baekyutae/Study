if location == len(temp) 이면 종료





```python
N = int(input())

member = list(map(int, input().split()))

def member_maker(member):
    member.sort(reverse=True)
    location = member[0]
    count = 0
    while location <= len(member):
        
        count += 1
        
        location += member[location]

    print(count)
```

## 구현상의 추가 문제(런타임 에러 가능)

location이 len(temp)보다 커질 수 있음, 그러면 다음 루프에서 member[location] 접근 시 바로 IndexError.

마찬가지로 while location <= len(member)면 location == len(member)인 순간에도 루프에 들어가서 내부에서 member[location]을 접근하면 바로 터짐

```python
# sort 할때 nlogn
# while문 n

# 시간복잡도 O(NlogN)

n = int(input())
member = list(map(int, input().split()))

def member_maker(member):
    member.sort()  # 오름차순
    location = 0   # 현재 보고 있는 인덱스
    count = 0      # 결성된 그룹 수
    cur = 0        # 현재 그룹에 모인 인원 수

    while location < len(member):   
        cur += 1                    # 한 명을 현재 그룹에 포함
        if cur >= member[location]: # 모인 인원수가 현재 그룹의 최대 공포 이상일때 그룹 확정
            count += 1
            cur = 0                 # 다음 그룹 새로 시작
        location += 1               # 다음 사람으로 이동

    print(count)

member_maker(member)


```

```python
# 같은로직 가독성 개선 한 버전

n = int(input())
member = list(map(int, input().split()))
member.sort()

groups = 0
cur = 0

for fear in member:
    cur += 1
    if cur >= fear:
        groups += 1
        cur = 0

print(groups)
```
