정수 x가 존재(1<= x <= 30,000)

사용가능한 연산 4가지

5로나누어 떨어지면 5로 나눔
3 으로 나누어 떨어지면 3으로 나눔
2로 나누어 떨어지면 2로 나눔
1을 뺀다

연산 4개를 활용해 1을 만들려고함

--------------------------

입력
정수 x

출력
연산을 사용하는 횟수의 최솟값

------------------

일단 큰수로 많이 나눌수록 최솟값에 가까워짐

해서 5부터 나누어 떨어지는지를 확인 5,3,2 순으로 안되면 1을빼서 나누어떨어지는 수를 찾아야함


일단 완전 탐색으로 푼다면

def make_one():
    연산 횟수를 셀 count = 0
    while x가 1일때 까지 반복

        x가 5로 나누어 떨어지면
            x = x//5
            count += 1
            처음으로
        x가 3로 나누어 떨어지면
            count += 1
            x = x//3
            처음으로
        x가 2로 나누어 떨어지면
            count += 1
            x = x//5
            처음으로

        나누어 떨어지는 수가 없으면
        x -= 1
        count += 1
