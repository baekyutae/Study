못생긴수: 오직 2,3,5 만 소인수로 가지는 수 ==  2,3,5를 약수로 가지는 합성수 

1은 못생긴 수라 가정

못생긴수가 1부터 오름차순으로 나열될때


n번째 못생긴 수를 찾는 프로그램을 작성

1<= n <= 1000

--------------

일단 못생긴 수 목록이 필요함

최대 1000번쨰 못생긴수 까지 알고있어야 문제의 주어진 범위 내에서 모든 못생긴수를 찾아 줄수 있음

가장 무식한 방법은

숫자 1부터 모든 자연수에 대해 못생긴수인지 확인해가며 카운트해

카운트가 1000이 되면 멈추는 방법이 있음

그리고 못생긴수의 특징이라면

2,3,5 중 최소 하나로 나누어 떨어지는 수임

반대로 생각하면

1에 2,3,5를 각각 곱하면 못생긴수 2,3,5가 나옴 못생긴 수 목록 [1,2,3,5]

2에 2,3,5를 각각 곱하면 못생긴수 4,6,10가 나옴 못생긴수 목록 [1,2,3,4,5,6,10]

3에 2,3,5를 각각 곱하면 못생긴수 6,9,15가 나옴 못생긴 수 목록 [1,2,3,4,5,6,9,10,15]


-----------------

앞서 존재하는 못생긴수에 2,3,5를 곱해가며 못생긴수 목록을 채워나가는 것 까진 했는데 그 이후 안풀림

그렇다면 **"세 개의 줄 서기(Queue)"**와 **"책갈피"** 개념으로 다시 설명해 볼게요. 이 방식이 가장 이해하기 쉽습니다.

핵심은 **`ugly`라는 리스트가 "재료"이자 동시에 "결과물"**이라는 점입니다.

---

### 1. 직관적 이해: 3개의 무한리필 가게

우리가 `ugly`라는 리스트를 만들어 나가는 과정을 **세 명의 요리사**가 음식을 만드는 것에 비유해 봅시다.

* **요리사 A (2배 담당):** 이미 만들어진 음식에 **2**를 곱해서 내놓음
* **요리사 B (3배 담당):** 이미 만들어진 음식에 **3**을 곱해서 내놓음
* **요리사 C (5배 담당):** 이미 만들어진 음식에 **5**를 곱해서 내놓음

처음에 접시엔 **`1`** 하나만 놓여 있습니다. (`ugly = [1]`)

이제 요리사들은 접시에 있는 음식을 **순서대로 가져와서** 자기 숫자를 곱해야 합니다.
그런데, 요리사마다 **작업 속도(현재 어디까지 처리했는지)**가 다릅니다.

* `i2`: 요리사 A가 "다음에 2를 곱해야 할 숫자가 `ugly` 리스트의 몇 번째 칸에 있는지" 기억하는 **책갈피**
* `i3`: 요리사 B의 **책갈피**
* `i5`: 요리사 C의 **책갈피**

---

### 2. 그림으로 보는 시뮬레이션

`ugly` 리스트가 칠판에 적혀있다고 상상하세요. 처음엔 `[1]`입니다.
세 명의 요리사(인덱스) 모두 처음엔 **첫 번째 숫자(1)**를 가리키고 있습니다.

#### [Round 1] 다음 숫자는 누구?

* 요리사 A: "난 1을 보고 있어. 내 다음 요리는 1 × 2 = **2**야." (`ugly[i2] * 2`)
* 요리사 B: "나도 1을 보고 있어. 내 다음 요리는 1 × 3 = **3**이야." (`ugly[i3] * 3`)
* 요리사 C: "나도 1을 보고 있어. 내 다음 요리는 1 × 5 = **5**야." (`ugly[i5] * 5`)

 제일 작은 **2**가 선택됩니다!

* `ugly` 리스트에 **2**를 적습니다.  `[1, 2]`
* **누가 2를 만들었죠? 요리사 A입니다.**
* 그럼 요리사 A에게 말합니다. "너는 1에다가 2 곱하는 건 끝났으니까, **책갈피 한 칸 옮겨서 다음 숫자 대기해.**"
* 이제 **`i2`는 1 증가**해서 `ugly` 리스트의 두 번째 숫자(방금 생긴 2)를 가리킵니다.
* (`i3`, `i5`는 아직 1을 처리 못 했으므로 그대로 첫 번째 숫자 1을 가리키고 있습니다.)

#### [Round 2] 다음 숫자는 누구?

* 요리사 A: (이제 두 번째 칸 '2'를 봄) "난 2를 보고 있어. 내 다음 요리는 2 × 2 = **4**야." (`ugly[i2] * 2`)
* *보세요, A는 이미 1x2는 끝냈으니 다음 타자인 2x2를 계산하는 겁니다.*


* 요리사 B: (아직 첫 번째 칸 '1'을 봄) "난 아직 1이야. 내 요리는 1 × 3 = **3**이야."
* 요리사 C: (아직 첫 번째 칸 '1'을 봄) "나도 아직 1이야. 내 요리는 1 × 5 = **5**야."

 이번엔 제일 작은 **3**이 선택됩니다!

* `ugly` 리스트에 **3**을 적습니다.  `[1, 2, 3]`
* **누가 3을 만들었죠? 요리사 B입니다.**
* 요리사 B에게 "너도 책갈피 한 칸 옮겨"라고 합니다. (`i3` 1 증가)

#### [Round 3] 여기가 중요합니다!

현재 상태:

* `ugly` 리스트: `[1, 2, 3]`
* `i2` 위치: 값 2를 가리킴 (인덱스 1)
* `i3` 위치: 값 2를 가리킴 (인덱스 1)  방금 증가해서 여기로 옴
* `i5` 위치: 값 1을 가리킴 (인덱스 0)  아직 한 번도 안 뽑힘

후보 계산:

* 요리사 A (`i2`): "난 2를 보고 있어. 2 × 2 = **4**"
* 요리사 B (`i3`): "난 2를 보고 있어. 2 × 3 = **6**" (1x3은 끝났으니 다음 타자 2x3)
* 요리사 C (`i5`): "난 아직 1이야. 1 × 5 = **5**"

 제일 작은 **4**가 선택됩니다.

* `ugly`에 4 추가.
* 요리사 A(`i2`)만 책갈피를 또 한 칸 옮깁니다. (이제 3을 가리킴)

### 3. 결론: 왜 `ugly[i2] * 2` 인가?

이 코드는 **"곱셈을 미리 다 해놓는 것"**이 아니라, **"각 배수(2배, 3배, 5배)별로 줄을 세워놓고, 맨 앞에서 하나씩 꺼내는 과정"**을 구현한 것입니다.

* `i2`는 **"2를 곱할 차례가 된 녀석이 줄 서 있는 위치"**입니다.
* 그래서 `next2 = ugly[i2] * 2`는 **"2배수 라인에서 지금 당장 나갈 수 있는 가장 작은 수"**를 의미하게 되는 것이죠.

이해를 돕기 위해 표로 정리하면 이렇습니다.

| 라운드 | ugly 리스트 (현재 상태) | i2 위치 (값) | i3 위치 (값) | i5 위치 (값) | 후보들 (2배/3배/5배) | 선택 |
| --- | --- | --- | --- | --- | --- | --- |
| 1 | `[1]` | idx 0 (1) | idx 0 (1) | idx 0 (1) | **1*2**, 1*3, 1*5 | **2** |
| 2 | `[1, 2]` | idx 1 (2) | idx 0 (1) | idx 0 (1) | 2*2, **1*3**, 1*5 | **3** |
| 3 | `[1, 2, 3]` | idx 1 (2) | idx 1 (2) | idx 0 (1) | **2*2**, 2*3, 1*5 | **4** |
| 4 | `[1, 2, 3, 4]` | idx 2 (3) | idx 1 (2) | idx 0 (1) | 3*2, 2*3, **1*5** | **5** |
| 5 | `[1, 2, 3, 4, 5]` | idx 2 (3) | idx 1 (2) | idx 1 (2) | **3*2**, **2*3**, 2*5 | **6** (중복!) |

이제 `i2`가 왜 필요한지, 왜 저 식을 쓰는지 조금 더 와닿으시나요?

네, 좋습니다! **Round 3 이후**부터 이어서 시뮬레이션을 쭉 보여드릴게요.
특히 **Round 5**에서 **값이 겹치는 경우(6)**가 나오는데, 이때 코드가 어떻게 동작하는지 보시면 완벽히 이해되실 겁니다.

### **현재 상황 (Round 3 종료 시점)**

* **만들어진 못생긴 수 리스트 (`ugly`):** `[1, 2, 3, 4]`
* **각 요리사의 현재 위치 (인덱스):**
* `i2` = 2 (세 번째 칸, 값 **3**을 가리킴)
* `i3` = 1 (두 번째 칸, 값 **2**를 가리킴)
* `i5` = 0 (첫 번째 칸, 값 **1**을 가리킴)

### **[Round 4] 5의 등장**

**1. 후보 선수 계산**

* **2배 팀 (`next2`):** `ugly[2] * 2` =  = **6**
* **3배 팀 (`next3`):** `ugly[1] * 3` =  = **6**
* **5배 팀 (`next5`):** `ugly[0] * 5` =  = **5**

**2. 선택 및 이동**

* 가장 작은 수는 **5**입니다. (`next5` 승리!)
* 리스트에 **5** 추가 → `[1, 2, 3, 4, 5]`
* **누가 만들었나?** 5배 팀(`i5`)입니다.
* **이동:** `i5`를 1 증가시킵니다. (이제 인덱스 1, 즉 값 **2**를 가리킴)

### **[Round 5] ★ 중요! 중복 발생 (6) ★**

이 라운드가 이 알고리즘의 핵심입니다.

**1. 후보 선수 계산**

* **2배 팀 (`next2`):** `ugly[2] * 2` =  = **6** (아직 선택 못 받아서 대기 중)
* **3배 팀 (`next3`):** `ugly[1] * 3` =  = **6** (아직 선택 못 받아서 대기 중)
* **5배 팀 (`next5`):** `ugly[1] * 5` =  = **10** (방금 5배 팀이 한 칸 이동했으므로, 값 2를 곱함)

**2. 선택 및 이동**

* 가장 작은 수는 **6**입니다.
* 그런데 **`next2`도 6이고, `next3`도 6**입니다!
* 리스트에 **6** 추가 → `[1, 2, 3, 4, 5, 6]`

**3. 포인터 이동 (여기가 핵심)**

* 코드의 `if`문은 둘 다 실행됩니다.
```python
if ugly[l] == next2:  # 6 == 6 (참)
    i2 += 1
if ugly[l] == next3:  # 6 == 6 (참)
    i3 += 1

```


* **2배 팀(`i2`):** "나 3에다가 2 곱해서 6 만들었어! 임무 완수!" → **한 칸 이동** (인덱스 3, 값 **4**를 가리킴)
* **3배 팀(`i3`):** "나도 2에다가 3 곱해서 6 만들었어! 임무 완수!" → **한 칸 이동** (인덱스 2, 값 **3**을 가리킴)
* **결과:** 6은 한 번만 적히고, 2배/3배 팀 모두 다음 숫자를 바라보게 됩니다.

### **최종 정리**

이제 패턴이 보이시나요?

1. **각 팀(2, 3, 5)은 자기만의 속도로 `ugly` 리스트를 훑어 나갑니다.**
2. `i2`, `i3`, `i5`는 각 팀이 **"어디까지 곱했는지"** 표시하는 책갈피입니다.
3. 매 라운드마다 각 팀이 내놓은 **가장 작은 결과물**만 채택되고, 채택된 팀만 **"다음 칸으로 이동"**합니다.
4. 만약 결과값이 같으면(6, 10 등), 해당되는 팀들이 **동시에 이동**합니다.

이 과정을 통해 순서가 꼬이지 않고 작은 수부터 차곡차곡 쌓이게 되는 것입니다.
