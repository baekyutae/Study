# 나눗셈을 하는 함수를 구현하세요.(순환소수 처리 포함)

입력=>출력 예시
5, 3 => 1.(6)
6, 2 => 3.0000
1, 3 => 0.(3)
 1, 7 => 0.(142857)

x,y = 나눌 대상 , 나누는 수 (정수)
() => 순환소수 묶음
순환 소수가 아닌경우 소숫점 네자리까지

입력값은 0 이상 정수
출력은 문자열
1.5139139139 이렇게 반복된다면 1.5(139), 비순환구간+순환구간 허용
xy < 10^9

잠긴 요구사항:

입력:입력값은 0 이상 정수

출력:출력은 문자열

예외(y=0 등):y가 0이면 0으로 나눌수 없습니다 출력

유한소수 포맷(반올림 규칙 포함): 5번째 소숫점 자리에서 반올림

순환소수 포맷(최소주기/비순환 허용 범위): 1.5139139139 이렇게 반복된다면 1.5(139), 비순환구간+순환구간 허용





문제 재정의(1~3문장):
두 0이상의 정수를 입력받아 나눗셈을 수행, 결과를 소숫점 넷째 자리까지 출력
다섯째 자리 이상일시 반올림, 순환소수부분은 전부다 적지 말고()로 묶음 또한 0.(142857) 나  0.129(79)와 같이 넷째 자리 까지 출력 제한을 받지 않음
=> 비순환 소수만 자리수 제약 존재
