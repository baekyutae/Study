배열 a,b 두배열은 n 개의 원소(모두 자연수)

최대 k번의 바꿔치기 연산

배열 a의 모든 원소의 합이 최대가 되도록

-------------
입력
1. n,k (n은 1이상 10만 이하, k는 0이상 n이하)
2. a의 원소들(천만보다 작은 자연수)
3. b의 원소들(천만보다 작은 자연수)
----------------
a의 작은 값들을 b의 더 큰값들로 순서대로 교체

a의 배열은 오름차순
b의 배열은 내림차순
정렬

index 0 부터 k-1 까지 두 배열 원소 교환
a의 원소보다 b의 원소가 큰경우에만 교환

a배열속 원소들의 합

def element_change(a,b,k):
    sorted_a = a 오름차순 정렬
    sorted_b = 내림차순 정렬
    for i in range(k):
        만약 a[i]보다 b[i]가 크면
            a[i] = b[i]
        그렇지 않으면
            교체 종료
    return sum(a)

case1. k가 0이라면?
그냥 sum해서 반환

case2. k = n 즉 전부  바꿔치기

1 2 3 4 7

6 6 5 5 5

case3. n,k 1이라면

a = 10 , b = 1000



```python
k=1
a=[1]
b=[5]

def element_change(a,b,k):
    sorted_a = sorted(a)
    sorted_b = sorted(b, reverse=True)
    for i in range(k):
        sorted_a[i] = sorted_b[i]
    return sum(sorted_a)

element_change(a,b,k)
```
