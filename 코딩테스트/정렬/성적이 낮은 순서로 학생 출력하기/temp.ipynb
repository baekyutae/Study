이름 - 성적
낮은 순서대로 이름을 출력

입력
1. 학생의 수 n (1<= n <= 100,000)
2 ~ n+1. 학생의 이름 문자열 a 와 성적에 해당하는 정수 b가 공백으로 구분되어 입력 (a의 길이와 성적은 100 이하의 자연수)

출력
모든 학생의 이름을 성적이 낮은 순으로 출력


----------------------------------

학생 이름과 점수를 짝지어 저장
점수를 기준으로 정렬 
이름을 출력



```python
n = int(input())
info = []
for _ in range(n):
    a, b =  input().split()
    info.append([a,b])

sorted_info = sorted(info, key = lambda x:x[1])

for i in sorted_info:
    print(i[0], end=' ')
```
