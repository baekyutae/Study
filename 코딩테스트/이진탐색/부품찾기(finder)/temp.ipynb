주문한 부품 번호를 하나씩 매장 부품 번호 배열에 대조해보며 존재 여부를 확인

------------------------

case1 n,m 모두 최대값
n이 백만 => 한번 이진탐색할때 약  20회
m도 백만 -> 백만개에 대해 존재여부 체크 => 2천만

```python
def finder(shop_list, order_list):
    shop_list = sorted(shop_list)
    result = []
    for i in order_list:
        target = i
        start = 0
        end = len(shop_list) -1
        check = False
        while start <= end:
            mid = (start+end) // 2
            if shop_list[mid] == target:
                check = True
                result.append("yes")
                break
            elif shop_list[mid] > target:
                end = mid - 1
            elif shop_list[mid] < target:
                start = mid + 1
                
        if check == False:
            result.append("no")
        

    return result

shop_list = [8,3,7,9,2]
order_list = [5,7,9]

finder(shop_list, order_list)
```
