```python
dduk_list = []
def dduk_maker(n:int, m:int, dduk_list:list):
    start = 0
    end = len(dduk_list) - 1
    dduk_list = sorted(dduk_list)
    while start <= end:
        mid = (start + end) // 2
        
        temp = []
        for i in dduk_list:
            if i == dduk_list[mid]:
                continue
            if i - dduk_list[mid] > 0:
                temp.append(i)
        if sum(temp) == m:
            return mid
        
        elif i - dduk_list[mid] > m:
            start = mid + 1
        else:
            end = mid - 1
# while 문 끝나기 직전의 start와 end 값
    
            
```
