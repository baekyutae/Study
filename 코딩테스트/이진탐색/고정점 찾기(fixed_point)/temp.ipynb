하나의 수열이 서로다른 원소 n개를 포함하고 있으며 오름차순으로 정렬

수열에서 고정점이 있다면 출력하는 프로그램을 작성 (고정점: 원소의 값이 인덱스 같은 원소)

고정점은 최대 1개만 존재 , 없으면 -1 출력

시간복잡도 logN 으로 설계하지 않으면 시간초과 판정을 받음

---------------------------------------------------

원소 n개가 있으면 인덱스의 범위는 0 ~ n-1

0 ~ n-1에서 하나씩 골라 고정점인지 이진탐색


def fixed_point(숫자배열):
    
    0 ~ n-1 까지의 정수에서 0 부터 하나씩 고름 (index)
        숫자배열 이진탐색
        시작점
        끝점
        while 인덱스 시작점과 끝점이 같을 떄 까지 탐색
            index랑  배열의 가운데 숫자(mid)가 같으면
            반복문 종료후 index 반환 (차피 고정점은 최대 1개)

            같지 않고 index가 mid보다 작으면
            왼쪽 탐색  시작
            같지 않고 index가 mid보다 크면
            오른쪽 탐색 시작

    while문 종료조건 까지 끝까지 다 돌았다면 고정점 없음
        -1 반환

case 1 : n이 1 원소가 0 => 0 반환

case 2 : n 이 백만 (원소가 백만개)

index가 백만개 

시간복잡도가 백만 * log백만 => 문제 발생 : 시간복잡도 초과

case 3: n은 5 배열은 [-15,-6,1,3,7]

```python
def fixed_point(n: int, num_list: list):
    for i in range(0,n):
        start = 0
        end = n-1
        
        while start <= end:
            mid = (start+end)//2

            if i == num_list[mid]:
                return i

            elif i >= num_list[mid]:
                start = mid + 1
            elif i <= num_list[mid]:
                end = mid - 1

    return -1
```

```python
n = 5
num_list = [-15,-6,1,3,7]
def fixed_point(n: int, num_list: list):
    
    start = 0
    end = n-1
        
    while start <= end:
        mid = (start+end)//2

        if mid == num_list[mid]:
            return mid

        elif mid >= num_list[mid]:
            start = mid + 1
        elif mid <= num_list[mid]:
            end = mid - 1

    return -1

fixed_point(n,num_list)
```
